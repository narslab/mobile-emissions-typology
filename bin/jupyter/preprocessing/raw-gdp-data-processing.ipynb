```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Data input

capita_ic_df = pd.read_excel('../../data/raw/lapi1123msa.xlsx') # modeshares (for commuting)
gdp_df = pd.read_csv('../../data/raw/Table.csv') # modeshares (for commuting)
```

```python
# Drop the first 7 rows and reset the index
capita_ic_df = capita_ic_df.drop(capita_ic_df.index[0:7]).reset_index(drop=True)

# Select only the first two columns
capita_ic_df = capita_ic_df.iloc[:, :2]

# Rename the columns
capita_ic_df.columns = ['MSA', 'Per Capita Personal Income 2020']

capita_ic_df = capita_ic_df[:-3]

capita_ic_df
```

```python
# Filter rows where Description is 'All industry total' or 'Government and government enterprises'
filtered_df = gdp_df[(gdp_df['Unnamed: 2'] == '1') | (gdp_df['Unnamed: 2'] == '83')]

# Convert numeric values to float where possible
filtered_df['Unnamed: 4'] = pd.to_numeric(filtered_df['Unnamed: 4'], errors='coerce')

# Group by Metropolitan Statistical Area and sum the values in Unnamed: 4
gdp_df = filtered_df.groupby('Unnamed: 1')['Unnamed: 4'].sum().reset_index()
gdp_df
```

```python
# Rename columns
gdp_df.columns = ['MSA', 'GDP']

# Remove "(Metropolitan Statistical Area)" from MSA column
gdp_df['MSA'] = gdp_df['MSA'].str.replace(r'(Metropolitan Statistical Area)', '')
```

```python
gdp_df.to_csv('../../data/tidy/msa-gdp.csv', index=False)
gdp_df
```
