```python
import geopandas as gpd

# Path to the MSA shapefile
msa_shapefile_path = '../../data/raw/cb_2018_us_cbsa_500k/cb_2018_us_cbsa_500k.shp'

# Read the shapefile into a GeoDataFrame
msa_gdf = gpd.read_file(msa_shapefile_path)

# Filter for Metropolitan Statistical Areas (MSA)
msa_gdf = msa_gdf[(msa_gdf['CBSAFP'].notnull()) & (msa_gdf['LSAD'] == 'M1')]
msa_gdf

# Reproject the data to EPSG:4326 (WGS 84) if not already in that projection
msa_gdf = msa_gdf.to_crs(epsg=4326)

# # Specify the path to save the shapefile and associated files
# shapefile_path = '../../data/tidy/msa_data_reprojected_only_msa.shp'

# # Save the shapefile along with all associated files (.shp, .shx, .dbf, .prj)
# msa_gdf.to_file(shapefile_path, driver='ESRI Shapefile')

#simplify the msa areas and regions
from shapely.geometry import Polygon

# Simplify the geometries in the GeoDataFrame
msa_gdf['geometry'] = msa_gdf['geometry'].apply(lambda geom: geom.simplify(tolerance=0.01, preserve_topology=True))

# Save the simplified file
msa_gdf.to_file("../../data/tidy/msa_data_simplified_msa.shp")
```
